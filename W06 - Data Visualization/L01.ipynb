```python
# Extract the feature data (sepal length, sepal width, petal length, petal width)   # thanks to chatgpt
data = iris.data   

# Plotting histogram for Sepal Length (feature index 0)
plt.figure(figsize=(8,6))
plt.hist(data[:, 0], bins=20, edgecolor='black', color='skyblue')

# Adding titles and labels
plt.title('Ciceklerin Turune Gore Ic Yaprak Uzunluklari')
plt.xlabel('Ic Yaprak Uzunlugu (cm)')
plt.ylabel('Cicek Sayisi')

# Show the plot
plt.show()
```

```python
iris = load_iris()

# Convert to a pandas DataFrame for easier manipulation
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Create a jointplot with KDE for Sepal Length vs Sepal Width
sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", data=iris_df, kind="kde", color="blue")

# Show the plot
plt.show()
```

```python

path = "/workspaces/AI01_H06/SampleSuperstore.csv"
df = pd.read_csv(path)

df
```

```python
num_variable = ['Sales', 'Quantity', 'Discount', 'Profit']

```

```python
fig, ax = plt.subplots(1,4)
plt.subplots_adjust(wspace=0.5)
for i, variable in enumerate(num_variable):
    sns.boxplot(data = df[variable], ax = ax[i] )
    ax[i].set_xlabel(num_variable[i])
    ax[i].set_xticklabels([])
plt.show()
```
