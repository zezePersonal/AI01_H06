```python
x = df[['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']]  # Yeni data tanimlayalim
y = df.Price
```

```python
# under-fitting, appropirate-fitting, over-fitting olaylari
# https://www.geeksforgeeks.org/underfitting-and-overfitting-in-machine-learning/
# img evernote'de var
# https://curiousily.com/posts/hackers-guide-to-fixing-underfitting-and-overfitting-models/

# Over-Fitting vs Under Fitting
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
```

```python
def get_MAE(max_leaf_nodes, train_x, test_x, train_y, test_y):  # Fonksiyon yazalim ki, fitting testi yapalim
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_x, train_y)
    p  = model.predict(test_x)
    mae = mean_absolute_error(test_y, p)
    return mae

```

```python
# simdi farkli parametreler gondererek max_leaf_nodes uzerinden bakalim

df.dropna(axis = 0)

for max_leaf_nodes in [5, 50, 500, 5000]:
    myMAE = get_MAE(max_leaf_nodes, train_x, test_x, train_y, test_y)
    print(f"Maksimum yaprak sayisi {max_leaf_nodes} iken hesaplanan MAE {myMAE}")

# zz: Bunun aslinda graph olarak cizilmesi gerekir
```
